**Clase Busqueda aleatoria**

```python
class BusquedaAleatoria:
  def __init__(self, xl, xu, yl, yu,err):
    self.__xl = xl
    self.__xu = xu
    self.__yl = yl
    self.__yu = yu
    self.__err = err
    self.__y = 0
    self.__x = 0
    self.__r = 0
    self.__maxf = 0
    self.__maxx = 0
    self.__maxy = 0
    self.__maximo = 0
    self.__lista = []
    self.__listaX = []
    self.__listaY = []
    self.__listaZ = []

  def __funcion(self):                                                                  #Evalúa f(x,y)
    f = self.__y - self.__x - 2*(self.__x**2) - 2*(self.__x)*(self.__y) - self.__y**2
    return f

  def __cal__r(self):                                                                    #Calcula el randonico r
    self.__r = random.random()

  def __cal_x_y(self):                                                                   #Calcula el valor x,y
    self.__cal__r()
    self.__x = self.__xl + (self.__xu - self.__xl)*self.__r
    self.__y = self.__yl + (self.__yu - self.__yl)*self.__r

  def get_maximo(self):
    return max(self.__lista)

  def get__listas(self):
    return self.__listaX,self.__listaY, self.__listaZ

  def Maximo(self):
    self.__maxf = -random.random()*1000
    for i in range(n):
      self.__cal_x_y()
      self.__listaX.append(self.__x)
      self.__listaY.append(self.__y)
      f = self.__funcion()
      self.__listaZ.append(f)
      if f > self.__maxf:
        self.__maxf = f
        self.__maxx = self.__x
        self.__maxy = self.__y
        self.__lista.append(f)
        print(i, " ",self.__x," ", self.__y," ", f)



```

```python
obj = BusquedaAleatoria(xl, xu, yl, yu,err)
obj.Maximo()
print("Maximo ",obj.get_maximo())
```
